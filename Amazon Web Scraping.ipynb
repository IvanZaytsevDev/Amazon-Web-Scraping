```python
# Import libraries
from bs4 import BeautifulSoup
import requests
import smtplib
import time
import datetime
```

```python
# Connect to Website and pull in data

URL = 'https://www.amazon.co.uk/Apple-iPhone-13-128GB-Blue/dp/B09G9HQHRH/ref=sr_1_3?crid=2VTVZ5H21WRK1&keywords=iphone+13&qid=1645613209&sprefix=iphone+1%2Caps%2C131&sr=8-3'

headers = {"User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/78.0.3904.108 Safari/537.36", "Accept-Encoding":"gzip, deflate", "Accept":"text/html,application/xhtml+xml,application/xml;q=0.9,*/*;q=0.8", "DNT":"1","Connection":"close", "Upgrade-Insecure-Requests":"1"}

page = requests.get(URL, headers=headers)

soup1 = BeautifulSoup(page.content, "html.parser")

soup2 = BeautifulSoup(soup1.prettify(), "html.parser")

title = soup2.find(id='productTitle').get_text()
price = soup2.find(id='corePrice_feature_div').get_text()
review = soup2.find(id='averageCustomerReviews').get_text()
print(title)
print(price)
print(review)
```

```python
# Clean up the data
review = review.strip()[0:12]
price = price.strip()[0:7]
title = title.strip()
print(title)
print(price)
print(review)
```

```python
# Create a Timestamp for your output to track when data was collected
today = datetime.date.today()
print(today)
```

```python
# Create CSV and write headers and data into the file
import csv
header = ['Title', 'Price', 'Review', 'Date']
data = [title, price, review, today]
with open('AmazonWebScraper.csv','w',newline='',encoding='UTF8') as f:
    writer = csv.writer(f)
    writer.writerow(header)
    writer.writerow(data)
```
